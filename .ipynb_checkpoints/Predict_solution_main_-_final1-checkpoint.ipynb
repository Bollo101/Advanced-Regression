```python
#importing libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
train_df['Valencia_pressure'] = train_df['Valencia_pressure'].fillna(train_df['Valencia_pressure'].median())
```

```python
train_df['Valencia_pressure'].isnull().sum()
```

```python
# Convert time from object to datetime
new_df = train_df.copy()
new_df['time']= pd.to_datetime(new_df['time'])
```

```python
new_df.info()
```

```python
new_df.head()
```

```python
new_df['Valencia_wind_deg'] = new_df['Valencia_wind_deg'].str.extract('(\d+)')
new_df['Seville_pressure'] = new_df['Seville_pressure'].str.extract('(\d+)')
new_df.head()
```

```python
new_df['Valencia_wind_deg'] = new_df['Valencia_wind_deg'].astype(float)
new_df['Seville_pressure'] = new_df['Seville_pressure'].astype(float)
```

```python
new_df.info()
```

```python
#new_df['Month'] = new_df['time'].dt.month
new_df['time_of_day'] = new_df['time'].dt.hour
new_df['Day'] = new_df['time'].dt.day
#new_df['Year'] = new_df['time'].dt.year
new_df.head()
```

```python
plt.figure(figsize = (40, 38))
cor = new_df.corr()
sns.heatmap(cor,annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
#Correlation with output variable
cor_target = abs(cor["load_shortfall_3h"])
#Selecting correlated features
relevant_features = cor_target[cor_target > 0.10]
print(relevant_features)
```

```python
drop_list2 = ['Madrid_temp','Seville_temp', 'Bilbao_temp', 
            'Barcelona_temp','Valencia_temp', 'time']    
```

```python
nw_df = new_df.copy()
```

```python
nw_df = nw_df.drop(columns=[col for col in nw_df if col in drop_list2])
```

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
X = nw_df.drop('load_shortfall_3h', axis=1)
y = nw_df['load_shortfall_3h']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=50)
```

```python
X_test.shape, y_test.shape
```

```python
X_train.shape, y_train.shape
```

#Random Forest Regression

```python
model2 = RandomForestRegressor()
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_test)
```

```python
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(y_test, y_pred2))
print('Coefficient of determination (R^2): %.2f'
     % r2_score(y_test, y_pred2))
print('RMSE:', mean_squared_error(y_test, y_pred2, squared=False))
```
