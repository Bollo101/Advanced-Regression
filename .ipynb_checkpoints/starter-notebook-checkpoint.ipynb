# Cleaning The Data for both Training and Testing sets
- we filled missing values of valencia pressure with the median
- we converted and normalised the Data for Valencia_wind_deg and Seville_pressure
- we converted time to its proper datetime format

```python
#cleaning the training Data
df['Valencia_pressure'] = df['Valencia_pressure'].fillna(df['Valencia_pressure'].median())
df['Valencia_wind_deg'] = df['Valencia_wind_deg'].str.extract('(\d+)')
df['Seville_pressure'] = df['Seville_pressure'].str.extract('(\d+)')
df['Valencia_wind_deg'] = df['Valencia_wind_deg'].astype(float)
df['Seville_pressure'] = df['Seville_pressure'].astype(float)
df['time'] = pd.to_datetime(df['time'])
```

```python
#Cleaning the Test Data
test_df['Valencia_pressure'] = test_df['Valencia_pressure'].fillna(test_df['Valencia_pressure'].median())
test_df['Valencia_wind_deg'] = test_df['Valencia_wind_deg'].str.extract('(\d+)')
test_df['Seville_pressure'] = test_df['Seville_pressure'].str.extract('(\d+)')
test_df['Valencia_wind_deg'] = test_df['Valencia_wind_deg'].astype(float)
test_df['Seville_pressure'] = test_df['Seville_pressure'].astype(float)
test_df['time']= pd.to_datetime(test_df['time'])
```

# Plotting a Time Series

## 5. Modelling

# Predictor vs Target Variables
- nw_df was set to conatin all relevant predictor variables from feature engeneering
- Target was set to load_shortfall_3h

```python
X = nw_df.drop('load_shortfall_3h', axis=1)
y = nw_df['load_shortfall_3h']
```

```python
#Creating the Test Train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=50)
X_train.shape, y_train.shape
```

# Selection of Models
- We decided to use a few different types of models for our data set and would be judging accuracy based on a MSE and RMSE 
## Models Used
- Linear - Simple Regression model 
- Ridge - Regression model using stadardisation and scaliing
- Random Forest - Regression model using random forking variables
- Decision Tree - Regression model using forking variables

### Training and Fitting the data on our selected models

```python
#Creating the Ridge Regression Model
ridge = Ridge()
ridge.fit(X_train, y_train)
```

```python
#Creating the linear Regression Model
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
#Creating the Random Forest Regression Model
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

```python
#Creating the Decision Tree Regression Model
regr_tree = DecisionTreeRegressor(max_depth=7,random_state=50)
regr_tree.fit(X_train, y_train)
```

```python
b0 = float(ridge.intercept_)
b1 = float(lm.intercept_)
```

```python
print("Intercept of ridge:", float(b0))
print("Intercept of linear:", float(b1))
```

# Creating Predictions for out models based on Training Data


```python
# get predictions for training data
train_lm = lm.predict(X_train)
train_ridge = ridge.predict(X_train)
train_forest =forest.predict(X_train)
train_decsion = regr_tree.predict(X_train)

print('Training MSE')
print('Random Forest:', metrics.mean_squared_error(y_train, train_forest).round(2))
print('Linear:', metrics.mean_squared_error(y_train, train_lm).round(2))
print('Ridge :', metrics.mean_squared_error(y_train, train_ridge).round(2))
print('Decision Tree :', metrics.mean_squared_error(y_train, train_decsion).round(2))
```

# Creating Predictions on Testing Data

```python
# get predictions for testing data
test_lm = lm.predict(X_test)
test_ridge = ridge.predict(X_test)
test_forest =forest.predict(X_test)
test_decsion = regr_tree.predict(X_test)


print('Testing MSE')
print('Random Forest:', metrics.mean_squared_error(y_test, test_forest).round(2))
print('Linear:', metrics.mean_squared_error(y_test, test_lm).round(2))
print('Ridge :', metrics.mean_squared_error(y_test, test_ridge).round(2))
print('Decision Tree :', metrics.mean_squared_error(y_test, test_decsion).round(2))
```

# Results
- While it seemed Decision Tree was the best model in the training phase, random forest pulled ahead in the testing phase so it will be our final model
- with a much lower RMSE our random forest model will be the best model to use in Predicting the load_shortfall for Spain

```python
print('RMSE for Testing Results')
print('Random Forest:', mean_squared_error(y_test, test_forest, squared=False).round(2))
print('Linear:', mean_squared_error(y_test, test_lm, squared=False).round(2))
print('Ridge:', mean_squared_error(y_test, test_ridge, squared=False).round(2))
print('Decision Tree:', mean_squared_error(y_test, test_decsion, squared=False).round(2))
```
